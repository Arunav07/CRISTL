```python
Image_array=[]
Image=tf.io.read_file("/content/drive/MyDrive/IDD_Detection/JPEGImages/frontFar/BLR-2018-03-22_17-39-26_2_frontFar/0000060.jpg")
im_decoded = tf.io.decode_jpeg(Image) 
img=tf.image.rgb_to_grayscale(im_decoded)
img = tf.image.central_crop(img, 0.85)
img = tf.image.resize(img, size=(256, 256))
img = np.asarray(img)
Image_array.append(img)
```
